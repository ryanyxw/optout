```python
filename1 = "out_val_oxford.csv"
filename2 = "out_val_nooxford.csv"
errorfile = "error.csv"
regex_oxford = ",(\s{0,1}\S+){1,3},\s{0,1}(and|or)\s{0,1}"
regex_nooxford = "(,\s?[^,\s]+){2,3} (and|or)\s?"


#,(\s?\S+){1,2},?\s?(and|or)\s
with open(errorfile, 'w') as csvfile0: 
    csvWriter0 = csv.writer(csvfile0)
    with open(filename1, 'w') as csvfile1:
        csvWriter1 = csv.writer(csvfile1)
        with open(filename2, 'w') as csvfile2:
            csvWriter2 = csv.writer(csvfile2)
        #     for i in tqdm(range(int(9e6))):
            for lineInd in tqdm(range(100)):
        #         sent_tokenize(re.sub(r'\n\n', ' ', json.loads(data[1])["text"]))
                tokenized_line = sent_tokenize(re.sub('\n', ' ', json.loads(data_val[lineInd])["text"]))
                csvStoreOxford = []
                csvStoreNoOxford = []
                error = []
                for sentence in tokenized_line:
        #             print(sentence)
        #             print(re.search(regex, sentence))
        #             break
                    isOxford = False
                    notOxford = False
                    if (re.search(regex_oxford, sentence) != None):
                        csvStoreOxford += [sentence]
                        isOxford = True
                    if (re.search(regex_nooxford, sentence) != None):
                        csvStoreNoOxford += [sentence]
                        notOxford = True
                    if (isOxford and notOxford):
                        error += [sentence]
        #         print(csvStore)
        #         break
                if (len(error) > 0):
                    csvWriter0.writerow(error)
                if (len(csvStoreOxford) > 0):
                    csvWriter1.writerow(csvStoreOxford)
        #             print(csvStore)
                if (len(csvStoreNoOxford) > 0):
                    csvWriter2.writerow(csvStoreNoOxford)
                



    
```

```python
#For the validation data generation

regex = ",([^,]){1,20}(,)?\s(and|or)\s"

outName = "val_oxford.csv"

with open(outName, 'w') as csvfile0: 
    csvWriter0 = csv.writer(csvfile0)
    csvWriter0.writerow(["lineInd", "sentence", "hasOxford", "index"])
    
    for lineInd in tqdm(range(3000)):
#         sent_tokenize(re.sub(r'\n\n', ' ', json.loads(data[1])["text"]))
        tokenized_line = sent_tokenize(re.sub('\n', ' ', json.loads(data_val[lineInd])["text"]))
        for sentence in tokenized_line:
            result = re.search(regex, sentence)
            if (result != None):
                hasOxford = result.group(2) == ','
#                 print(result)
#                 print(result.group(2))
                isAnd = (result.group(3) == "and")
                offSet = 6 if isAnd else 5
        
                
                index = result.span()[1] - offSet
                
                #This is so that our index is always at the position that is supposed to be the comma (for non oxford, this is the space)
                if (not hasOxford):
                    index+= 1
#                 print(result.group(3))
#                 print(sentence[index])
    
                    
                csvWriter0.writerow([lineInd, sentence, hasOxford, index])
        
```
